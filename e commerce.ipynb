Trực quan hóa

Vòng đời khách hàng

```python
#Loại bỏ CustomerID == 0 để tránh lặp nhiễu
cus = data[data['CustomerID'] != '0']
cus = cus.groupby(['First Purchase Date', 'Monthdiff'])['CustomerID'].nunique().unstack()
#không biểu diễn 12/2010 vì thiếu dữ liệu kh cũ trc đó
cus = cus[cus.index != '2010-12']
cus
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=cus, cmap = 'RdBu', annot= True, fmt=".0f")
plt.title(label= 'Số lượng KH trong năm 2011', size = 16)
plt.yticks(rotation = 0)
plt.show()
```

```python
#Biểu diễn theo tỉ lệ
for column in sorted(cus.columns, reverse=True):
    cus[column] = cus[column]/cus[0]
cus
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=cus, annot= True, fmt=".0%", cmap= 'RdBu')
plt.yticks(rotation = 0)
plt.grid(False)
plt.title(label= 'Tỉ lệ khách hàng quay lại năm 2011', size = 16)
plt.ylabel('')
plt.xlabel('Thứ tự tháng')
plt.show()
```

Cohort theo doanh thu

```python
cus = data[data['CustomerID'] != '0']
cus = cus.groupby(['First Purchase Date', 'Monthdiff'])['revenue'].sum()/1000
cus = cus.unstack()
cus = cus[cus.index != '2010-12']
cus
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=cus, annot= True, fmt=".0f", cmap= 'RdBu')
plt.yticks(rotation = 0)
plt.grid(False)
plt.title(label= 'Doanh thu (nghìn đô)', size = 16)
plt.ylabel('')
plt.xlabel('Thứ tự tháng')
plt.show()
```

Cycle Life Value

```python
cus = data[~data['CustomerID'].isin(['0'])]
cus = cus.groupby(['First Purchase Date', 'Monthdiff']).agg({'revenue':'sum', 'CustomerID':'nunique'})
cus['mean'] = cus['revenue']/cus['CustomerID']
# cus = cus.unstack()

cus = cus.unstack()['mean']
cus = cus[cus.index != '2010-12']
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=cus, annot= True, fmt=".0f", cmap= 'RdBu')
plt.yticks(rotation = 0)
plt.grid(False)
plt.title(label= 'Trung bình doanh thu của từng khách (đô)', size = 16)
plt.ylabel('')
plt.xlabel('Thứ tự tháng')
plt.show()
```

Phân tích RFM 

```python
RFM_cus = data.groupby('CustomerID').agg({'InvoiceNo':'nunique', 'revenue':'sum', 'InvoiceDate': 'max' })
RFM_cus = RFM_cus[RFM_cus.index != '0']
RFM_cus.sort_values(by = 'InvoiceDate')
# RFM_cus['Recency'] = (pd.to_datetime('2011-12-1') - RFM_cus['InvoiceDate']).dt.days
RFM_cus['Recency'] = (RFM_cus['InvoiceDate'].max() - RFM_cus['InvoiceDate']).dt.days
RFM_cus.describe()
```

```python
RFM_cus['Frequency_value'] = pd.qcut(RFM_cus['InvoiceNo'], 5, labels= False, duplicates='drop') +1
RFM_cus['Recency_value'] = pd.qcut(RFM_cus['Recency'], 5, labels= False, duplicates='drop') +1
RFM_cus['Monetary_value'] = pd.qcut(RFM_cus['revenue'], 5, labels= False, duplicates='drop') +1
RFM_cus['Recency_value'] = 6 - RFM_cus['Recency_value'] #đảo ngược xếp hạng 
RFM_cus['RFM_total'] = RFM_cus['Recency_value'].astype('str') + RFM_cus['Frequency_value'].astype('str') + RFM_cus['Monetary_value'].astype('str')
RFM_cus['RFM_total'] = RFM_cus['RFM_total'].astype('int')
RFM_cus.reset_index(inplace= True)
RFM_cus
```
